```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xd=[1,2,3,4]
yd=[2,4,6,8]
```

```python
w=1.0
```

```python
def forward(x):
    return x*w
```

```python
def loss(x,y):
    y_=forward(x)
    return (y-y_)*(y-y_)
```

```python
loss_list=[]
weight_list=[]
```

```python
for w in np.arange(0.1,4.0,0.1):
    
    tot_loss=0
    for x,y in zip(xd,yd):
        l=loss(x,y)
        tot_loss += l
    
    loss_list.append(tot_loss/len(xd))
    weight_list.append(w)
```

```python
plt.plot(weight_list,loss_list)
plt.xlabel('weights')
plt.ylabel('loss')
plt.show()
```
