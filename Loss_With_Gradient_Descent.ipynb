```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xd=[1.0,2.0,3.0]
yd=[2.0,4.0,6.0]
```

```python
plt.plot(xd,yd)
plt.show()
```

```python
w=0.01
```

```python
def forward(x):
    return x*w
```

```python
def loss(x,y):
    y_=forward(x)
    return (y-y_)*(y-y_)
```

```python
def gradient(x,y):
    return 2*x*(x*w-y)
```

```python
loss_list=[]
weight_list=[]
```

```python
for epoch in range(10):
    
    for x,y in zip(xd,yd):
        g=gradient(x,y)
        w=w-0.01*g
        l=loss(x,y)
        if(epoch%10==0):
            loss_list.append(l)
            weight_list.append(w)
```

```python
plt.plot(weight_list,loss_list)
plt.show()
```
